```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import psutil
```

```python
def matrix_multiply(constant_matrix, num_threads):
    random_matrix = np.random.rand(1000, 1000)
    StartTime = time.time()
    np.dot(random_matrix, constant_matrix)
    EndTime = time.time()
    TimeTaken = EndTime - StartTime
    CPUUsage = psutil.cpu_percent()
    print(f"CPU usage: {CPUUsage}%")
    return TimeTaken
```

```python
ConstMatrix = np.random.rand(1000, 1000)
NoThreads = []
for i in range(8):
    NoThreads.append(i + 1)
print(NoThreads)
```

```python
TimeTakenList = []
for t in NoThreads:
    TimeTaken = matrix_multiply(ConstMatrix, t)
    TimeTakenList.append(TimeTaken)
ResultDf = pd.DataFrame({'Threads': NoThreads, 'Time Taken (Sec)': TimeTakenList})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(NoThreads, TimeTakenList, marker='o')
plt.xlabel('Number of Threads')
plt.ylabel('Time Taken')
plt.grid(True)
plt.show()
```

```python
print("\nResult Table:")
ResultDf=ResultDf.to_string(index=False, justify='center')
print(ResultDf)
```
